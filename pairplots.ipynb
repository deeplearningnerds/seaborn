## Import Packages

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

The seaborn function load_datasets() loads a seaborn dataset into a pandas dataframe. Let's load the iris dataset.

```python
iris_df = sns.load_dataset("iris")

iris_df.head()
```

## Pairplot

Now, we would like to visualize the data with a pairplot. A pairplot visualizes the disribution of each numeric variable and the relation to the other variables. Therefore a pairplots can be very helpful  for exploratory data analysis.

```python
#pairplot
sns.pairplot(data=iris_df)

plt.show()
```

### Using further Parameters

```python
#pairplot
sns.pairplot(data=iris_df, vars=["sepal_length","sepal_width"])

plt.show()
```

```python
#pairplot
sns.pairplot(data=iris_df, hue="species")

plt.show()
```

```python
#pairplot
sns.pairplot(data=iris_df, hue="sepal_width")

plt.show()
```

```python
#pairplot
sns.pairplot(data=iris_df, kind="kde")

plt.show()
```

```python
#pairplot
sns.pairplot(data=iris_df, corner=True)

plt.show()
```

```python
#pairplot
sns.pairplot(data=iris_df, height=1.2)

plt.show()
```
