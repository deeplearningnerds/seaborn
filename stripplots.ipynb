## Import Packages

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

The seaborn function load_datasets() loads a seaborn dataset into a pandas dataframe. Let's load the tips dataset.

```python
tips_df = sns.load_dataset("tips")

tips_df.head()
```

## Strippplot

Now, we would like to visualize the tips data with a stripplot. 

Stripplots are a special form of scatter plots and can be used to display the distribution of a numeric variable by a specific category. 

By default, the points are randomly distributed along the categorical axis. This makes the points overlap less and clusters of data points become more visible. 

In order to generate a stripplot, we use the stripplot() function from Seaborn. 

```python
sns.stripplot(data=tips_df)

plt.show()
```

### Using further Parameters

#### Using the x parameter

In our first example we want to plot the tip. To do this, we set the x parameter to 'tip'.

```python
sns.stripplot(data=tips_df, x="tip")

plt.show()
```

#### Using the x and y parameter

Now, we want to plot the distribution of the tips for each day. In addition to the x parameter, we use the y parameter and set it to 'day'. 

```python
sns.stripplot(data=tips_df, x="tip", y="day")

plt.show()
```

#### Using the hue parameter

Now, we want to highlight the time. In order to do this, we use the hue parameter and set it to 'time'.

```python
sns.stripplot(data=tips_df, x="tip", y="day", hue="time")

plt.show()
```

#### Using the dodge parameter

Now, we want to split the distribution for each categery. In order to do this, we use the dodge parameter and set it to True.

```python
sns.stripplot(data=tips_df, x="tip", y="day", hue ="time", 
              dodge=True)

plt.show()
```

#### Using styling parameters

Now, let's modify the styling of the plot. We use stars as markers and choose a different color palette.

```python
sns.stripplot(data=tips_df, x="tip", y="day", hue ="time", 
              dodge=True, marker="*", size=22, alpha=0.3, 
              palette= "mako")

plt.show()
```
