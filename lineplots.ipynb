## Import Packages

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

The seaborn function load_datasets() loads a seaborn dataset into a pandas dataframe. Let's load the flights dataset. This dataset contains the number of passengers over a period of 10 years.

```python
flights_df = sns.load_dataset("flights")

flights_df.head()
```

## Lineplot

Now, we would like to visualize the data with a lineplot.

### Number of Passengers at December over the years

Let's analyze the number of passengers in December and how it has developed over the years.

```python
december_flights_df = flights_df.query("month == 'Dec'")

december_flights_df.head()
```

```python
#figure style
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(12, 7))

#lineplot
sns.lineplot(x='year', y='passengers', data=december_flights_df)

plt.show()
```

### Number of Passengers at each month over the years

Let's analyze the development of the number of passengers for all months over the years. To do this, we consider our DataFrame without filters. The parameter 'hue' specifies the grouping.

```python
#figure style
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(12, 7))

#lineplot
sns.lineplot(x='year', y='passengers', hue='month', data=flights_df)

plt.show()
```

### Number of passengers over the years on an annual basis

Let's analyze the development of the number of passengers  on an annual basis. In this case, there are several values for each year. Seaborn aggregates the values for each year and the plot shows the mean and the 95% confidence interval by default.

```python
#figure style
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(12, 7))

#lineplot
sns.lineplot(x='year', y='passengers', data=flights_df)

plt.show()
```
