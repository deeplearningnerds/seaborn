## Import Packages

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

The seaborn function load_dataset() loads a seaborn dataset into a pandas dataframe. Let's load the penguins dataset. 

```python
penguins_df = sns.load_dataset("penguins")

penguins_df.head()
```

## Jointplot

Now, we would like to visualize the data with a jointplot. We consider the flipper length and the bill length of the pengiuns dataset. We want to analyze the distribution of the two variables.

Let's use the function jointplot() und try different values for the kind parameter.

```python
sns.jointplot(data=penguins_df, x='flipper_length_mm', y='bill_length_mm',
              color = 'darkcyan',
              kind = 'scatter')

plt.show()
```

```python
sns.jointplot(data=penguins_df, x='flipper_length_mm', y='bill_length_mm',
              color = 'darkcyan',
              kind = 'hex')

plt.show()
```

```python
sns.jointplot(data=penguins_df, x='flipper_length_mm', y='bill_length_mm',
              color = 'darkcyan',
              kind = 'kde')

plt.show()
```

Now, let's add the hue parameter to analyze the data for each species.

```python
sns.jointplot(data=penguins_df, x='flipper_length_mm', y='bill_length_mm', 
              palette = 'dark:cyan',
              hue = 'species',
              kind = 'scatter')

plt.show()
```

```python
sns.jointplot(data=penguins_df, x='flipper_length_mm', y='bill_length_mm',
              palette = 'dark:cyan',
              hue = 'species', 
              kind = 'kde')

plt.show()
```
