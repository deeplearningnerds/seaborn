## Import Packages

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

The seaborn function load_datasets() loads a seaborn dataset into a pandas dataframe. Let's load the penguins dataset.

```python
penguins_df = sns.load_dataset("penguins")

penguins_df.head()
```

## Swarmplot

Now, we would like to visualize the data with a swarmplot

```python
sns.swarmplot(data=penguins_df)

plt.show()
```

### Using further Parameters

#### Using the x parameter

In our first example we want to plot the flipper length. To do this, we set the x parameter to 'flipper_lenth_mm'.

```python
sns.swarmplot(data=penguins_df, x="flipper_length_mm")

plt.show()
```

#### Using the x and y parameter

Now, we want to plot the distribution of the flipper length for each species. In addition to the x parameter, we use the y parameter and set it to 'species'. 

```python
sns.swarmplot(data=penguins_df, x="flipper_length_mm", y="species")

plt.show()
```

#### Using the size parameter

Now, we want to decrease the size of the markers. In order to do this, we use the size parameter and set it to 3.

```python
sns.swarmplot(data=penguins_df, x="flipper_length_mm", y="species", 
              size=3)

plt.show()
```

#### Using the hue parameter

Now, we want to highlight the sex of the penguins. In order to do this, we use the hue parameter and set it to 'sex'.

```python
sns.swarmplot(data=penguins_df, x="flipper_length_mm", y="species", 
              size=3, hue ="sex")

plt.show()
```

#### Using the dodge parameter

Now, we want to split the distribution for each categery into separate swarms. In order to do this, we use the dodge parameter and set it to True.

```python
sns.swarmplot(data=penguins_df, x="flipper_length_mm", y="species", 
              size=3, hue="sex", dodge=True)

plt.show()
```
