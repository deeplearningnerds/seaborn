## Import Packages

```python
import seaborn as sns
```

## Overview of Seaborn Datasets

Seaborn provides some example datasets that can be easily loaded into Python.

```python
sns.get_dataset_names()
```

## Load Dataset

The seaborn function load_datasets() loads the dataset into a pandas dataframe. Let's have a look at some example datasets.

### Flights dataset

```python
flights_df = sns.load_dataset("flights")

flights_df.head()
```

### Peguins dataset

```python
penguins_df = sns.load_dataset("penguins")

penguins_df.head()
```

### Taxis dataset

```python
taxis_df = sns.load_dataset("taxis")

taxis_df.head()
```

### Tips dataset

```python
tips_df = sns.load_dataset("tips")

tips_df.head()
```

### Planets dataset

```python
planets_df = sns.load_dataset("planets")

planets_df.head()
```
