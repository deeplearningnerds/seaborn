## Import Packages

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

The seaborn function load_datasets() loads a seaborn dataset into a pandas dataframe. Let's load the penguins dataset.

```python
penguins_df = sns.load_dataset("penguins")

penguins_df.head()
```

## Violinplot

Now, we would like to visualize the data with a violinplot. A violinplot is similar to a boxplot, but in addition the underlying distribution for each category is visualized.


```python
#figure style
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(12, 7))

#violinplot
sns.violinplot(x='species', y='body_mass_g', data=penguins_df, kind="kde")

plt.show()
```

### Changing the Figure Style

```python
#figure style
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(8, 4))

#plot violins
sns.violinplot(x='species', y='body_mass_g', data=penguins_df)

plt.show()
```

```python
#figure style
sns.set_style('white')
fig, ax = plt.subplots(figsize=(8, 4))

#plot violins
sns.violinplot(x='species', y='body_mass_g', data=penguins_df)

plt.show()
```

```python
#figure style
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(8, 4))

#plot violins
sns.violinplot(x='species', y='body_mass_g', data=penguins_df)

plt.show()
```

```python
#figure style
sns.set_style('dark')
fig, ax = plt.subplots(figsize=(8, 4))

#plot violins
sns.violinplot(x='species', y='body_mass_g', data=penguins_df)

plt.show()
```

```python
#figure style
sns.set_style('ticks')
fig, ax = plt.subplots(figsize=(8, 4))

#plot violins
sns.violinplot(x='species', y='body_mass_g', data=penguins_df)

plt.show()
```
